```python
def empty_hand() -> bool:
    """
    determines if the bot's hotbar slot #1 is empty
    """
    if bot.inventory.slots[36] is None or bot.inventory.slots[36].stackSize == 1:  # 36 is hot bar slot 1
        return True
    else:
        return False


def empty_inv() -> bool:
    """
    determines if the bot's inventory slot #1 is empty
    """
    for i, slot in enumerate(bot.inventory.slots):
        if slot is not None and bot.inventory.slots[i].stackSize != 1:
            return False
    return True


def gift_inv():
    for i, slot in enumerate(bot.inventory.slots):
        if slot is not None and bot.inventory.slots[i].stackSize != 1:
            bot.moveSlotItem(i, 36)
            break


def set_creative():
    bot.chat("/gamemode 1")


def fly_right():
    bot.creative.startFlying()
    pos = bot.entity.position.offset(1, 0, 0)
    bot.creative.flyTo(pos)
    bot.physics.gravity = 0.08


def fly_left():
    bot.creative.startFlying()
    pos = bot.entity.position.offset(-1, 0, 0)
    bot.creative.flyTo(pos)

    # bot.waitForTicks(5)
    source_position = bot.entity.position.offset(0, -1.5, 0)
    source_block = bot.blockAt(source_position)

    faceVector = {'x': 0, 'y': 1, 'z': 0}
    bot.placeBlock(source_block, faceVector)


def fly_down():
    bot.creative.startFlying()
    pos = bot.entity.position.offset(0, -1, 0)
    bot.creative.flyTo(pos)


def fly_up():
    faceVector = {'x': 0, 'y': 1, 'z': 0}
    bot.creative.stopFlying()
    bot.creative.startFlying()

    pos = bot.entity.position.offset(0, 1, 0)
    bot.creative.flyTo(pos)

    # bot.waitForTicks(5)
    source_position = bot.entity.position.offset(0, -1.5, 0)
    source_block = bot.blockAt(source_position)

    bot.placeBlock(source_block, faceVector)
#     print("Y pos:", bot.entity.position.y)


def set_survival():
    bot.chat("/gamemode 0")
    bot.creative.stopFlying()


def break_block():
    pos = bot.entity.position.offset(0, -1, 0)
    blockUnder = bot.blockAt(pos)
    if bot.canDigBlock(blockUnder):
        bot.chat(f"I'm breaking the '{blockUnder.name}' block underneath")
        # The start=True parameter means to immediately invoke the function underneath
        # If left blank, you can start it with the `start()` function later on.
        try:
            @AsyncTask(start=True)
            def break_block(task):
                bot.dig(blockUnder)

            bot.chat('I started digging!')
        except Exception as e:
            bot.chat(f"I had an error {e}")
    else:
        bot.chat(f"I can't break the '{blockUnder.name}' block underneath")


def clear_inventory():
    bot.chat("/clear")


def build():
    # Starts jumping
    bot.setControlState('jump', True)
    # Look at block underneath
    while True:
        position = bot.entity.position.offset(0, -0.5, 0)
        block = bot.blockAt(position)
        if block.name == 'air':
            bot.chat(str(bot.entity.position))
            break
    bot.chat(str(block.name))

    source_position = bot.entity.position.offset(0, -1.5, 0)
    source_block = bot.blockAt(source_position)

    # Place block
    if empty_hand():
        gift_inv()
    if empty_inv():
        bot.setControlState('jump', False)
        return bot.chat("No More Blocks to Build Up!")

    faceVector = {'x': 0, 'y': 1, 'z': 0}
    bot.placeBlock(source_block, faceVector)

    # Stop Jumping
    bot.setControlState('jump', False)
```

```python
"""
Tasklist:
- Figure out how many unique blocks there are
- Sort them out into order, Ex. Black_wool: 1, Wool: 2, Orange_wool: 3, etc
- Create a Function where it checks the height difference
- Create a Function where it checks for surrounding mobs

"""

@On(bot, 'chat')
def breakListener(this, user, message, *args):
    # print(bot.inventory.slots)
    if user and (user != bot.username):
        if PREFIX + 'clear' in message:
            clear_inventory()
        if PREFIX + 'quickbar' in message:
            bot.equip(37, 'hand')
        if PREFIX + 'creative' in message:
            set_creative()
        if PREFIX + 'survival' in message:
            set_survival()
        if PREFIX + 'stop' in message:
            off(bot, 'chat', breakListener)
            
```
